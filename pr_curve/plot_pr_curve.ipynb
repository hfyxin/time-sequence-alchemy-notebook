```
import pickle
import matplotlib.pyplot as plt

TITLE_SIZE = 16
LABEL_SIZE = 14
TICK_SIZE = 12
LEGEND_SIZE = 16
```

```
# load curve data

path = r'/content/drive/My Drive/time_sequence_alchemy/pr_curve/filter_120s/'
pkl_fns = {
  'RNN': 'RNN.pkl',
  'CNN': 'CNN.pkl',
  'Ts-RNN': 'Ts-RNN.pkl',
  'Ts-CNN': 'Ts-CNN.pkl', 
}
pr_data = {}

for name, fn in pkl_fns.items():
  with open(path+fn, 'rb') as f:
    pr_data[name] = pickle.load(f)
```

Single plot of the balanced dataset

```
label_txt = []

plt.figure(figsize=(7,6))
for label, data in pr_data.items():
  label_txt.append(label)
  plt.plot(data['pr'], data['rc'])
  break
plt.legend(label_txt, fontsize=LABEL_SIZE)
plt.xlim((-0.05, 1.05))
plt.xlabel('Precision', fontsize=LABEL_SIZE)
plt.ylabel('Recall', fontsize=LABEL_SIZE)
plt.xticks(fontsize=TICK_SIZE)
plt.yticks(fontsize=TICK_SIZE)
plt.grid()
```

Single plot of the imbalanced dataset

```
label_txt = []

plt.figure(figsize=(7,6))
for label, data in pr_data.items():
  label_txt.append(label)
  plt.plot(data['pr_cvrt'], data['rc_cvrt'])
plt.legend(label_txt, fontsize=LABEL_SIZE)
plt.xlim((-0.05, 1.05))
plt.xlabel('Precision', fontsize=LABEL_SIZE)
plt.ylabel('Recall', fontsize=LABEL_SIZE)
plt.xticks(fontsize=TICK_SIZE)
plt.yticks(fontsize=TICK_SIZE)
plt.grid()
```

Dual plot.

```
label_txt = []

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
for label, data in pr_data.items():
  label_txt.append(label)
  plt.plot(data['pr'], data['rc'])
plt.legend(label_txt, fontsize=LABEL_SIZE)
plt.xlim((-0.05, 1.05))
plt.xlabel('Precision', fontsize=LABEL_SIZE)
plt.ylabel('Recall', fontsize=LABEL_SIZE)
plt.title('P-R curve', fontsize=TITLE_SIZE)
plt.xticks(fontsize=TICK_SIZE)
plt.yticks(fontsize=TICK_SIZE)
plt.grid()

plt.subplot(1,2,2)
for label, data in pr_data.items():
  label_txt.append(label)
  plt.plot(data['pr_cvrt'], data['rc_cvrt'])
plt.legend(label_txt, fontsize=LABEL_SIZE)
plt.xlim((-0.05, 1.05))
plt.xlabel('Precision', fontsize=LABEL_SIZE)
plt.ylabel('Recall', fontsize=LABEL_SIZE)
plt.title('Equivalent P-R curve', fontsize=TITLE_SIZE)
plt.xticks(fontsize=TICK_SIZE)
plt.yticks(fontsize=TICK_SIZE)
plt.grid()
```
